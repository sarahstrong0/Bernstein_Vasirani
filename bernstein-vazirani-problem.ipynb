###The Bernstein-Vazirani problem:
Input: a function f : {0, 1}<sup>n</sup> → {0, 1}.<br>
Assumption: f(x) = a × x + b.<br>

Output: a,b.<br>
Notation: {0, 1}<sup>n</sup> is the set of bit strings of length n, a is an unknown bit string of length n, × is inner product mod 2, + is addition mod 2, and b is an unknown single bit.

```python
import numpy as np

def test_f(x):
  a = [1,0,1,0]
  b = 0
  return (np.dot(a,x) % 2)

def find_a_b(f, n):
  # find b by calculating f at zero
  zero_vec = [0] * n
  b = f(zero_vec)

  # a and x must be the same size
  a = [0] * n
  # find a by calculating f with all zero bits except ith bit = 1
  # f([0,0,0,1]), f([0,0,1,0]), etc

  for i in range(n):
    a_test = [0] * n
    a_test[i] = 1
    a_at_i = f(a_test)
    # store each bit of a
    # subtract b and take mod from each bit
    a[i] = (a_at_i - b) % 2

  return a,b

find_a_b(test_f,4)
```
